```python
import pygame
pygame.init()

screen = pygame.display.set_mode((600, 600))
pygame.display.set_caption("Tic Tac Toe")

WHITE = (255, 255, 255)
BLACK = (0, 0, 0)

size = 200
font = pygame.font.Font(None, 100)

board = [["", "", ""],
         ["", "", ""],
         ["", "", ""]]

player = "X"
run = True
winner = None
game_over = False

def check_winner(b):
    
    for r in range(3):
        if b[r][0] == b[r][1] == b[r][2] != "":
            return b[r][0]

    for c in range(3):
        if b[0][c] == b[1][c] == b[2][c] != "":
            return b[0][c]

    if b[0][0] == b[1][1] == b[2][2] != "":
        return b[0][0]
    if b[0][2] == b[1][1] == b[2][0] != "":
        return b[0][2]

    return None


while run:
    screen.fill(WHITE)

    # Draw grid
    pygame.draw.line(screen, BLACK, (200, 0), (200, 600), 5)
    pygame.draw.line(screen, BLACK, (400, 0), (400, 600), 5)
    pygame.draw.line(screen, BLACK, (0, 200), (600, 200), 5)
    pygame.draw.line(screen, BLACK, (0, 400), (600, 400), 5)

    for r in range(3):
        for c in range(3):
            if board[r][c] != "":
                text = font.render(board[r][c], True, BLACK)
                screen.blit(text, (c * size + 70, r * size + 50))

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False

        if not game_over and event.type == pygame.MOUSEBUTTONDOWN:
            x, y = pygame.mouse.get_pos()
            row = y // 200
            col = x // 200

            if board[row][col] == "":
                board[row][col] = player

                winner = check_winner(board)
                if winner:
                    game_over = True

               
                player = "O" if player == "X" else "X"

    if winner:
        win_font = pygame.font.Font(None, 80)
        win_text = win_font.render(f"{winner} wins!", True, (255, 0, 0))
        screen.blit(win_text, (170, 260))

    pygame.display.update()

pygame.quit()
```
