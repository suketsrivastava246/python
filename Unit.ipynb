```python
while True:
    print("\n---Unit Conversion-----" )
    print("choose conversion :")
    print("1. Length Conversion")
    print("2. Weight Conversion")
    print("3.Exit")

    main = int(input("choose category(1-3):"))

    if main == 1:
        print("\nLength Conversion Mneu")
        print("1. Meter to Kilometer")
        print("2. Kilometer to Meter")
        print("3. Centimeter to Meter")
        print("3. Meter to Centimeter")

        ch=int(input("Enter your choice (1-4): "))

        if ch == 1:
            m = float(input("Enter meters:"))
            print("Kilometer:" , m / 1000)

        elif ch == 2:
            km = float(input("Enter Kilometer"))
            print("Meter:",km * 1000)

        elif ch == 3:
            cm = float(input("Enter Centimeter"))
            print("Meter:",cm / 100)

        elif ch == 4:
            m = float(input("Enter meters: "))
            print("Centimeters:", m * 100)

        else:
            print("Invalid length option!")

    elif main ==2:
        print("exit")
        break
        
    else:
        print("Invalid Choice!")

    stop = input("\nDo you want to continue? (yes/no):").lower()
    if stop == "no":
        print ("exit...stop")
        break
```
