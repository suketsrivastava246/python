```python
# print("------MOVIE TICKET BOOKING SYSTEM--------")

movies = {
    1: {"name": "Avengers", "price": 150},
    2: {"name": "KGF", "price": 180},
    3: {"name": "Inception", "price": 200}
}

while True:
    print("\nAvailable Movie: ")
    for key, value in movies.items():
        print(f"{key},{value['name']} - {value['price']} per ticket")

    choice = int(input("\nChoose movie number: "))

    if choice not in movies:
        print("Invalid movie selection")
        continue

    tickets = int(input("Enter number of tickets: "))
    
    if tickets <= 0:
        print("Invalid number of tickets!")
        continue

    selected_movie = movies[choice]
    total = tickets * selected_movie["price"]

    print("\n---- Booking Summary ----")
    print("Movie Name : ", selected_movie["name"])
    print("Tickets    : ", tickets)
    print("Price      : ", selected_movie["price"])
    print("Total Cost : ", total)

    confirm = input("\nConfirm booking? (yes/no): ").lower()
    
    if confirm == "yes":
        print("Booking Confirmed! Enjoy the movie")
    else:
        print("Booking Cancelled!")

    again = input("\nBook another ticket? (yes/no): ").lower()
    if again == "no":
        print("Thank you for using Movie Booking System!")
        break
```
